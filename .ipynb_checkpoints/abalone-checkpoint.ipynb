# DATA-PREPROCESSING

```python
#fixing/handling categerical features
df['Sex']=df['Sex'].map({'F':0,'M':1,'I':2})
df.head()
```

```python
#lets see if any outliers are present
df.plot(kind='box',subplots=True,layout=(2,5),figsize=(10,6))
```

as we can see there are outliers in every column

# lets try to remove the outliers

```python
x = df.drop('Rings',axis=1)
```

```python
y = df['Rings']
```

# using zscsore method to remove outliers

```python
from scipy.stats import zscore
z=np.abs(zscore(x))
df_new=df[(z<3).all(axis=1)]
df_new
```

```python
print(df.shape)
print(df_new.shape)
```

```python
loss = 4177-4027
print(f'the numbers of rows lost are {loss}')
```

```python
datalosspercentage=(((4177-4027)/4177)*100)
print(f'the data loss percentage {datalosspercentage}')
```

```python
#lets check again if any outliers are present
df_new.plot(kind='box',subplots=True,layout=(2,5),figsize=(10,6))
```

we can see still some outliers are present

# lets use IQR method to remove other outliers

```python
Q1=df_new.quantile(0.25)
Q3=df_new.quantile(0.75)
IQR=Q3 - Q1

df_1=df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
#lets check again if any outliers are present
df_1.plot(kind='box',subplots=True,layout=(2,5),figsize=(10,6))
```

now we can see there are no outliers present

```python
df_1.skew()
```

we can see skewness in LENGTH,DIAMETER,HEIGHT

```python
from sklearn.preprocessing import PowerTransformer
scaler = PowerTransformer(method='yeo-johnson')
features=['Length','Diameter','Age']
```

```python
df_new[features] = scaler.fit_transform(df_new[features].values)
```

```python
x = df_new.drop("Rings",axis=1)
y = df_new["Rings"]
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
x = pd.DataFrame(sc.fit_transform(x), columns=x.columns)
```

# lets find the best accuracy state

```python
#importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30, random_state =i)
    mod = DecisionTreeRegressor()
    mod.fit(x_train, y_train)
    pred = mod.predict(x_test)
    acc=r2_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ",maxAccu," on Random_state ",maxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=maxRS)
```

```python
from sklearn.tree import DecisionTreeRegressor
DTR=DecisionTreeRegressor()
DTR.fit(x_train,y_train)
pred=DTR.predict(x_test)
print('R2_Score:',r2_score(y_test,pred))
```

# Cross validation score

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(DTR,x,y,cv=5).mean())
```

# Hyper Parameter Tuning

```python
Final_mod=DecisionTreeRegressor(criterion='mse',max_depth=5,max_features='auto',random_state=1)
Final_mod.fit(x_train,y_train)
pred=Final_mod.predict(x_test)
acc=r2_score(y_test, pred)

print(acc*100)
```

# Save the Model

```python
import joblib
joblib.dump(Final_mod,"Abalone.pkl")
```
