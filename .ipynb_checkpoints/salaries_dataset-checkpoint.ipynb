```python
df = df.drop(np.where(df['salary'] > 200000)[0])
```

```python
df = df.drop(np.where(df['yrs.service'] >40)[0])
```

```python
df = pd.get_dummies(df, columns = ['rank', 'discipline','sex'])
```

```python
df
```

```python
from scipy.stats import zscore
from scipy import stats
import numpy as np
z=np.abs(zscore(df))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
newdf = df[(z < 3).all(axis=1)]
```

```python
#dependent and indipendent variable
independent=newdf.iloc[:,:-1]
dependent=newdf.iloc[:,-1]
```

```python
from sklearn.preprocessing import power_transform
x=power_transform(independent,method='yeo-johnson')
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
lr=LinearRegression()
```

```python
#splitting the data into training and testing
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=.33,random_state=47)
```

```python
lr.fit(x_train,y_train)
```

```python
predlr=lr.predict(x_test)
```

```python
lr.score(x_train,y_train)
```
