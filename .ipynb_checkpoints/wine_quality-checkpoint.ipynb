```python
y
```

```python
x=df.drop('quality',axis=1)
x
```

```python
plt.figure(figsize=(20,25),facecolor='white')
plotnum=1
for col in x:
    if plotnum<=11:
        plt.subplot(5,5,plotnum)
        sns.distplot(x[col])
        plt.xlabel(col,fontsize=20)
    plotnum+=1
plt.show()
```

'sulphates', 'fixed acidity', 'volatile acidity', 'residual sugar', 'chlorides','free sulfur dioxide', 'total sulfur dioxide' ,'sulphates', 'alcohol' are all right skewed distributions.

```python
y = y.apply(lambda y_val : 1 if y_val>=7 else 0)
y
```

```python
sns.countplot(y, palette="Set1")
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
test_accuracy = accuracy_score(pred, y_test)
test_accuracy
```
